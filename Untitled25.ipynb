```python
# Label Encoding 
from sklearn.preprocessing import LabelEncoder

# Label Encoder object
le = LabelEncoder()

# Apply Label Encoding on 'education' column
df['education'] = le.fit_transform(df['education'])

# Check result
df[['education']].head()
```

```python
#  One-Hot Encoding 
# Select categorical columns for One-Hot Encoding
categorical_cols = [
    'workclass',
    'marital.status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'native.country'
]

# Apply One-Hot Encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Check dataframe
df.head()
```

```python
#Scale numerical features using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Numerical columns 
num_cols = [
    'age',
    'fnlwgt',
    'education.num',
    'capital.gain',
    'capital.loss',
    'hours.per.week'
]

# Apply scaling
df[num_cols] = scaler.fit_transform(df[num_cols])

# Check result
df[num_cols].head()
```

# Comparing model readiness before and after scaling.
Before scaling, numerical columns had different value ranges. Some columns had very large values, while others had small values.

After scaling, all numerical columns were converted to a similar scale with mean around 0 and standard deviation around 1.

This makes the dataset more suitable and fair for machine learning models.

# Distance-based algorithms
Algorithms like KNN and K-means use distances between points.
If features have very different scales, large values dominate the distance.
Scaling fixes this problem.

# Gradient-based algorithms
Algorithms like Linear Regression, Logistic Regression, Neural Networks use gradients to learn.
If features are on different scales, learning can be slow or unstable.
Scaling makes learning faster and more stable.

# Tree-based algorithms
Algorithms like Decision Trees, Random Forests are not affected by scaling.

# Summary
Scaling helps models perform better and ensures that no feature dominates just because of its scale.
